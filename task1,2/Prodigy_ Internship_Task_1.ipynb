```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('E:/prodigy inturn/task1/train.csv')

# Create a bar chart for Gender distribution
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(gender_counts.index, gender_counts.values, color='lightblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()


# Create a histogram for Age distribution
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()



```

# Conculsion 

Age Distribution:  The histogram of passenger ages shows that most passengers are in their 20s to 30s, with fewer older and younger passengers. There are some gaps in the age data due to missing values.

Gender Distribution: The bar chart for gender distribution shows that there were significantly more male passengers than female passengers on the Titanic. This can be useful for further analysis, especially when combined with other variables, such as survival rate.

By visualizing the distribution of age and gender, we can better understand the demographic structure of the Titanic passengers, which is key when performing deeper analyses such as survival prediction based on these features.
